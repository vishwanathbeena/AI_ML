```python
iris_setosa = iris.loc[iris["species"] == "Iris-setosa"];
iris_virginica = iris.loc[iris["species"] == "Iris-virginica"];
iris_versicolor = iris.loc[iris["species"] == "Iris-versicolor"];
#print(iris_setosa)

sample_array=np.array([1,2,2,2,2,2,2,3,3,4,4,4,4,4,5,6,7,8,9,9,9,9,9,9,9,9])

print(sample_array.size)
print('array_elements are')
print(sample_array)
counts_np, bin_edges_np = np.histogram(sample_array, bins=3, 
                                 density = False)

db = np.diff(bin_edges_np)

print("counts are")
print(counts_np)
print("************")
print("bin_edges are")
print(bin_edges_np)
print('diff bin edges is')
print(db)
print("************")
pdf = counts_np/(sum(counts_np))
print('pdf is')
print(pdf)
cdf = np.cumsum(pdf)
plt.plot(bin_edges_np[1:],pdf);
plt.plot(bin_edges_np[1:],cdf);
plt.show()
plt.close()
plt.figure('sns_distplot')
sns.distplot(sample_array,label='numpy plot')
plt.show()
plt.close()



print('*********** WITH TRUE**********')
counts_np, bin_edges_np = np.histogram(sample_array, bins=3, 
                                 density = True)

print("counts are")
print(counts_np)
print("************")
print("bin_edges are")
print(bin_edges_np)
print("************")

pdf = counts_np/(sum(counts_np))
print('pdf is')
print(pdf)
print('*******************ENDS HERE****************')

counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, 
                                 density = True)
print("counts are")
print(counts)
print("************")
print("bin_edges are")
print(bin_edges)
print("************")
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.show()
```
